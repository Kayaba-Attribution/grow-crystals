```python
import pandas as pd 
df = pd.read_csv("../results/sweep_results.csv", header=None)#, index_col=)
df.columns = ['harmonic','data_size','d_model','batch_size','parallelogram','variances']

print(df.head(5))
print(df.info())
```

```python
no_harmonic_df = df[df['harmonic'] == 0]
import matplotlib.pyplot as plt 
plt.subplot(121)
plt.plot(no_harmonic_df['data_size'], no_harmonic_df['parallelogram'])
plt.subplot(122)
plt.plot(no_harmonic_df['data_size'], no_harmonic_df['variances'])

```

```python
harmonic_df = df[df['harmonic'] == 1]
import matplotlib.pyplot as plt 
plt.subplot(121)
plt.plot(harmonic_df['data_size'], harmonic_df['parallelogram'])
plt.subplot(122)
plt.plot(harmonic_df['data_size'], harmonic_df['variances'])
```
